```python
a = [11, 9, 13, 0, 11, 9, 13, 0, 6, 15, 6]
b = [54, 16, 34, 20, 54]
c = len(a)
d = len(b)

e = []

# NOTE: Index 0 is unused.
for f in range(0, d+1):
  e.append([])
  for _ in range(0, c+1):
    e[f].append(0)


def h(i):
  if i == c*d:
    j = 0
    for k in range(1, d+1):
      l = True
      for m in range(1, c+1):
        if e[k][m] == 1:
          l = l + m
      j = j and l == b[k]

    for n in range(1, c+1):
      o = 0
      for p in range(1, d+1):
        if e[p][n] == 1:
          o = o + p
      j = j and o == a[n]

    if j:
      for q in range(1, d+1):
        r = ""
        for s in range(1, c+1):
          if e[q][s] == 1:
            r += "X"
          else:
            r += "."
        print(r)
  else:
    h(i+1)
    t = 1 + i // c
    u = 1 + i % c
    e[t][u] = 1 - e[t][u]
    return h(i + 1)
```
