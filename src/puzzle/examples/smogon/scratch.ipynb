```python
from itertools import product
from string import ascii_uppercase as ABC
from time import time


def _insert_newlines(string, every=128):
    return "\n".join(string[i:i + every] for i in range(0, len(string), every))

def round1(grid, n_solutions=1):
    solution_counter = [n_solutions]
    t0 = time()
    S = grid.count("\n")
    seen = set()
    LENGHT = S * S
    N = int(grid[0])
    TOTAL = N * S
    grid = "".join(grid[1:].split())
    possibles = tuple(set(i for i in range(LENGHT) if grid[i] is l)
                    for l in ABC[:S])
    regions = sorted(range(S), key=lambda x: grid.count(ABC[:S][x]))
    adjacent = []
    for a in range(LENGHT):
        cols = [0]
        rows = [0]
        if a % S > 0:
            cols.append(-1)
        if a % S < S - 1:
            cols.append(1)
        if a // S > 0:
            rows.append(-S)
        if a // S < S - 1:
            rows.append(S)
        adjacent.append(set(a + sum(i) for i in product(cols, rows)))
    cs = tuple(set(range(i, LENGHT, S)) for i in range(S))
    rs = (0,) * S + tuple(set(range(i * S, i * S + S)) for i in range(S))
    cr = tuple((i % S, i // S + S) for i in range(LENGHT))

    def solve(grid, impossibles, colrows, count=0):
        if count == TOTAL:
            solution_counter[0] -= 1
            if not solution_counter[0]:
                return grid
            print('Discarding:', grid)
            return None
        for a in possibles[regions[count // N]] - impossibles:
            c, r = cr[a]
            tryimpossibles = impossibles | adjacent[a]
            trycolrows = colrows[:]
            trycolrows[c] += 1
            trycolrows[r] += 1
            if trycolrows[c] == N:
                tryimpossibles.update(cs[c])
            if trycolrows[r] == N:
                tryimpossibles.update(rs[r])
            s = solve(grid + (a,), tryimpossibles, trycolrows, count + 1)
            if s:
                break
        else:
            return None
        return s
    s = solve((), set(), bytearray(S + S))
    if s:
        s = "".join("*" if i in s else "." for i in range(LENGHT))
        print(_insert_newlines(s, S))
    print("Took", time() - t0, "seconds")
```

```python
from data import data
with open(data.project_path('src/puzzle/examples/smogon/he_comes.txt'), 'rb') as f:
  he_comes = f.read()
```

```python
def _binary(n):
  return bin(n)[2:].rjust(8, '0')

def process(text):
  i = 0
  result = []
  while i < len(text):
    n = text[i]
    binary = _binary(n)
    offset = binary.find('0') or 1
    result.append(text[i:i+offset])
    i += offset
  return result
```

```python
import unicodedata

processed = process(he_comes)
MAP = {
  # MORSE.
  b'\xe2\x83\xa8': '...',
  b'\xcc\xa6': ',',
  b'\xcc\xa3': '.',
  b'\xcc\xb1': '-',
  b'\xcc\xb3': '--',
  b'\xcc\xa4': '..',
  b'\xe2\x83\x9c': '....',
  # Semaphore?
  b'\xcc\xad': '^',
  b'\xcd\x94': '<',
  b'\xcd\x95': '>',
  b'\xcc\xac': 'v',
  # Mayan Numerals?
  # ???
  # Small letters
  b'\xcd\xa3': 'a',
  b'\xcd\xa8': 'c',
  b'\xcd\xa9': 'd',
  b'\xcd\xa4': 'e',
  b'\xe1\xb7\x9a': 'g',
  b'\xcd\xaa': 'h',
  b'\xcd\xa5': 'i',
  b'\xe1\xb7\x9d': 'l',
  b'\xcd\xab': 'm',
  b'\xe1\xb7\xa0': 'n',
  b'\xcd\xa6': 'o',
  b'\xe1\xb7\xae': 'p',
  b'\xcd\xac': 'r',
  b'\xe1\xb7\xa4': 's',
  b'\xcd\xad': 't',
  b'\xcd\xa7': 'u',
  b'\xe1\xb7\xb1': 'w',
  b'\xcd\xaf': 'x',
  b'\xe1\xb7\xa6': 'z',
  b'\xcc\x93': '\'',
  # Glagolitic script
  b'\xf0\x9e\x80\x8d': '/k/',
  b'\xf0\x9e\x80\xa0': '/ə/',
  b'\xf0\x9e\x80\x93': '/r/',
  b'\xf0\x9e\x80\x85': '/ɛ/',
  b'\xf0\x9e\x80\x89': '/i/',
  b'\xf0\x9e\x80\x8f': '/m/',
  b'\xf0\x9e\x80\x82': '/ʋ/',
  b'\xf0\x9e\x80\x90': '/n/',
  b'\xf0\x9e\x80\x95': '/t/',
  b'\xf0\x9e\x80\x91': '/ɔ/',
  b'\xf0\x9e\x80\x94': '/s/',
}
PIGPEN = {
  b'\xcc\xa1': 'aj',  # (top left)
  b'\xcc\xba': 'bk',  # (top)
  b'\xcc\xa2': 'cl',  # L
  b'\xcc\xb9': 'dm',  # ]
  b'\xcc\xbb': 'en',  # []
  b'\xcc\x9c': 'fo',  # [
  b'\xcd\x89': 'gp',  # (bottom left)
  b'\xcc\xac': 'sw',  # v
  b'\xcd\x95': 'tx',  # >
  b'\xcd\x94': 'uy',  # <
}
all_mapped = []
i = 0

def is_pigpen(n):
  return 262 <= n <= 296

while i < len(processed):
  c = processed[i]
  if c.isalnum():
    mapped = c.decode()
  elif is_pigpen(i) and c in PIGPEN:
    next_c = processed[i+1]
    mapped = PIGPEN[c][next_c == b'\xcc\xa3']
    i += next_c == b'\xcc\xa3'
  else:
    mapped = MAP.get(c, '?')
  if not c.isalnum():
    print('%d\t%s\t%s\t%s (%s)\t%s' % (
      i, c, mapped, c.decode(), unicodedata.name(c.decode()), '\t'.join(_binary(b) for b in c)))
  #if mapped is None:
  #  break
  all_mapped.append(mapped)
  i += 1
print(''.join(all_mapped))
```
