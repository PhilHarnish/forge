```python
import forge
from puzzle.puzzlepedia import puzzlepedia

puzzle = puzzlepedia.parse("""
name in {Beth, Charles, David, Frank, Jessica, Karen, Taylor}
start in {800, 810, 820, 830, 840, 850, 860}
route in {NorthCW, NorthCCW, SouthCW, SouthCCW, RoundCW, RoundCCW, SeaCaves}

# Setup:
# TODO: This shouldn't be necessary.
for n in name:
  sum(n[s] for s in start) == 1

#1: Lists different times. (In setup.)
# Players start at 8:00, 8:10, 8:30, 8:40, 8:50, 9:00 ("8:60").

#2: Describes routes. (In setup.)

#3: Describes distances. (Examples below.)
def t_pond(n):
  n.NorthCW or n.NorthCCW or n.SouthCW or n.SouthCCW
  return n.start + (
    (n.NorthCW * (10 + 30 + 25 + 10)) +
    (n.NorthCCW * (10 + 10)) +
    (n.SouthCW * (10 + 10)) +
    (n.SouthCCW * (10 + 45 + 15 + 10))
  )

def t_well(n):
  return n.start + (
    (n.NorthCW * (10 + 30 + 25)) +
    (n.NorthCCW * (10 + 10 + 10)) +
    (n.SouthCW * (10 + 10 + 10)) +
    (n.SouthCCW * (10 + 45 + 15)) +
    (n.RoundCW * (10 + 30 + 25)) +
    (n.RoundCCW * (10 + 45 + 15)) +
    (n.SeaCaves * (10 + 10 + 10))
  )

def t_lookout_point(n):
  n.SouthCW or n.SouthCCW or n.RoundCW or n.RoundCCW
  return n.start + (
    (n.SouthCW * (10 + 10 + 10 + 15)) +
    (n.SouthCCW * (10 + 45)) +
    (n.RoundCW * (10 + 30 + 25 + 15)) +
    (n.RoundCCW * (10 + 45))
  )

def t_return(n):
  return n.start + (
    (n.NorthCW * (10 + 30 + 25 + 10 + 10 + 10)) +
    (n.NorthCCW * (10 + 10 + 10 + 25 + 30 + 10)) +
    (n.SouthCW * (10 + 10 + 10 + 15 + 45 + 10)) +
    (n.SouthCCW * (10 + 45 + 15 + 10 + 10 + 10)) +
    (n.RoundCW * (10 + 30 + 25 + 15 + 45 + 10)) +
    (n.RoundCCW * (10 + 45 + 15 + 25 + 30 + 10))
  )

#4:
t_well(Frank) > t_well(Charles)
t_well(Frank) < t_well(Beth)

#5a: David left 10 minutes after Jessica...
David.start == Jessica.start + 10
#5b: ...and passed each other @ Lookout Point...
t_lookout_point(David) == t_lookout_point(Jessica)
#5c: ...going opposite directions.
# David passing through Lookout Point E->W == Jessica W->E
(David.SouthCW or David.RoundCW) == (Jessica.SouthCCW or Jessica.RoundCCW)

#6a: David and Karen were at pond at same time...
t_pond(David) == t_pond(Karen)
#6b: ...going opposite directions.
# David passing through pond E->W == Karen W->E.
(David.NorthCW or David.SouthCCW) == (Karen.NorthCW or Karen.SouthCW)

#7:
Taylor == 820
Taylor == SeaCaves

#8:
t_return(Frank) == t_return(Beth) + 10

print('Well times:')
print('Beth', t_well(Beth))
print('Charles', t_well(Charles))
print('David', t_well(David))
print('Frank', t_well(Frank))
print('Jessica', t_well(Jessica))
print('Karen', t_well(Karen))
print('Taylor', t_well(Taylor))
""")
```
