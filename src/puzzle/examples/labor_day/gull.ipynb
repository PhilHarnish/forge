```python
def walk(seek_set, acc, remaining):
  if remaining <= 0:
    yield ' '.join(acc)
    return
  seek_set.set_length(remaining)
  for result, weight in trie.walk(seek_set):
    if weight < 5e4:
      break
    acc.append(result)
    yield from walk(seek_set[result:], acc, remaining - len(result))
    acc.pop()

walk(seek_set, [], 10)
```
