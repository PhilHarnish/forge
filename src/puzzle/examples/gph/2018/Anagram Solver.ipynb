```python
def compute_map(before, after):
  return {
    k: v for k, v in zip(before, after)
  }
```

```python
def find_first(src, s):
  result = []
  for c in s:
    try:
      result.append(src.index(c))
    except:
      result.append(-1)
  return result
```
