```python
import collections

Processed = collections.namedtuple('Processed', ['phrases', 'categories'])

def process(examples):
  phrases = collections.Counter()
  categories = collections.defaultdict(list)
  processed = Processed(phrases, categories)
  for example in examples:
    for line in example.split('\n'):
      phrase, suffix = line.split(' of the ')
      phrase = phrase.split()
      for p in phrase:
        phrases[p] += 1
      category, _, level = suffix.split()
      level = int(level.strip(')'))
      categories[category].append((phrase, level))
  return processed
```
